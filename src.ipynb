```python
import pandas as pd
import torch
import numpy as np
```

```python
data = pd.read_csv("college_student_placement_dataset.csv")
```

```python
data.head()
```

```python
data.drop(labels="College_ID",inplace=True,axis=1)
```

```python
data.head()
```

```python
data["Internship_Experience"].replace(to_replace=["No","Yes"],value=[0,1],
                                      inplace=True)
```

```python
data.head()
```

```python
data["Placement"].replace(to_replace=["No","Yes"],value=[0,1],inplace=True)
```

```python
data.head()
```

```python
X = torch.tensor(np.array(data.iloc[:,0:-1]),dtype=torch.float32)
y = np.array(data.iloc[:,-1])
y = y.reshape(y.shape[0],1)
y = torch.tensor(y, dtype=torch.float32)
```

```python
print(X)
```

```python
torch.cuda.is_available()
```

```python
class binary_classification_nn(torch.nn.Module):

    def __init__(self):
        super().__init__()

        self.first_hidden_layer = torch.nn.Linear(in_features=8,out_features=8)
        self.first_hidden_layer_function = torch.nn.ReLU()
        self.second_hidden_layer = torch.nn.Linear(in_features=8,out_features=8)
        self.second_hidden_layer_function = torch.nn.ReLU()
        self.output_layer = torch.nn.Linear(in_features=8,out_features=1)

    def forward(self,X):

        z2_vec = self.first_hidden_layer(X)
        h2_vec = self.first_hidden_layer_function(z2_vec)

        z3_vec = self.second_hidden_layer(h2_vec)
        h3_vec = self.second_hidden_layer_function(z3_vec)

        y_hat = self.output_layer(h3_vec)

        return y_hat
```

```python
model = binary_classification_nn()
```

```python
for params in model.parameters():
    print(params)
```

```python
loss_function = torch.nn.BCEWithLogitsLoss()
optimizer = torch.optim.RMSprop(params=model.parameters())
epoch = 0
tol = 10**(-6)

while True:

    initial_y_hat = model(X)
    initial_loss = loss_function(initial_y_hat,y)
    optimizer.zero_grad()
    initial_loss.backward()
    optimizer.step()
    final_y_hat = model(X)
    final_loss = loss_function(final_y_hat,y)

    epoch += 1
    if abs(final_loss.item() - initial_loss.item()) < tol:
        break

    print("Epoch #{}, BCE Loss Function Value = {}".format(epoch,initial_loss))
```
