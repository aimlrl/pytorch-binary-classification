```python
class binary_classification_nn(torch.nn.Module):

    def __init__(self):
        super().__init__()

        self.first_hidden_layer = torch.nn.Linear(in_features=8,out_features=8)
        self.first_hidden_layer_function = torch.nn.ReLU()
        self.second_hidden_layer = torch.nn.Linear(in_features=8,out_features=8)
        self.second_hidden_layer_function = torch.nn.ReLU()
        self.output_layer = torch.nn.Linear(in_features=8,out_features=1)
        self.output_layer_function = torch.nn.Sigmoid()

    def forward(self,X):

        z2_vec = self.first_hidden_layer(X)
        h2_vec = self.first_hidden_layer_function(z2_vec)

        z3_vec = self.second_hidden_layer(h2_vec)
        h3_vec = self.second_hidden_layer_function(z3_vec)

        z4_vec = self.output_layer(h3_vec)
        y_hat = self.output_layer_function(z4_vec)

        return y_hat
```

```python
model = binary_classification_nn()
```

```python
loss_function = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.RMSprop(params=model.parameters())
```
